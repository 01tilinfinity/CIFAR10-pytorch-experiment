##**Train loss**

```python
from tables.file import TransactionG
path = '/content/drive/MyDrive/experiment'
file_list = os.listdir(path)
trainloss_csv = [file for file in file_list if file.endswith('trainloss.csv')]
print(trainloss_csv)
```

```python
trainloss_csv[3],trainloss_csv[4] = trainloss_csv[4],trainloss_csv[3]
trainloss_csv.reverse()
print(trainloss_csv)
```

```python
average_loss=list()

for files in trainloss_csv:
  files = pd.read_csv(files)
  files = pd.DataFrame(files)
  value = files['Value']
  value = value.tolist()
  sum = 0
  for loss in value:
    sum = sum + loss
  total_loss = sum / 90
  average_loss.append(total_loss)

print(average_loss)
```

```python
batch_size =[8,16,32,64,128,256]
losses = average_loss
x = np.arange(6)
plt.xticks(x,batch_size)
plt.bar(x,losses)
plt.xlabel('Batch Size')
plt.ylabel('Loss')
plt.title('Train Loss')

plt.show()
```
